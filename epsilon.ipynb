```julia
## using LinearAlgebra
using PyCall
using PyPlot
using LaTeXStrings
using Distributions
```

```julia
wp_ag=3.7 #eV
wp_au=5.8
wp_al=15
ima=1im;
τ=1e-7;
e=1.06217e-19;
h=6.626068e-34;
x=e/h;
```

```julia
function ε(ω,ωp) 
    
    return 1-(ωp^2)/((ω^2)*(1+ima/(ω*τ))) 
end

function ε1(ω,ωp)
return real(ε(ω,ωp))
end

function ε2(ω,ωp)
return imag(ε(ω,ωp))
end
```

$Ag$, $Au$, $Al$ 
Obtener la curva de respuesta para esos elementos, buscar la frecuencia de plasmón

```julia
ω=2.7:0.1:5;


eAg=ε2.(ω*x,wp_ag*x)
#eAu=ε1.(ω*x,wp_au*x)
#eAl=ε1.(ω*x,wp_al*x)


plot(ω*x,eAg)
#plot(ω*x,eAu)
#plot(ω*x,eAl)

xlabel(L"\omega (Hz)")
ylabel(L"\epsilon (ω)")
```
